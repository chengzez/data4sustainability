```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates
df_resident = pd.read_csv("Residential.csv")
df_industrial = pd.read_csv("Industrial.csv")
df_commercial = pd.read_csv("Commercial.csv")

df_resident= df_resident[df_resident['YYYYMM'] >= 200600]
df_industrial= df_industrial[df_industrial['YYYYMM'] >= 200600]
df_commercial= df_commercial[df_commercial['YYYYMM'] >= 200600]
```

```python
columnlist =[1, 2 ,3 ,10, 12 ,13]
datalist = []
df_residentfinal = df_resident[df_resident['Column_Order'] == 13].drop(columns=['Value','Column_Order','Description']).reset_index()
for ind in columnlist:
    df1 = df_resident[df_resident['Column_Order'] == ind]
    df_residentfinal[str(df1['Description'].iloc[0])] = list(df1['Value'])
    
df_residentfinal['date'] = [int(i) for i in df_residentfinal['YYYYMM']]
df_residentfinal['Total Energy'] = [float(i) for i in df_residentfinal['Total Energy Consumed by the Residential Sector']]
df_residentfinal 
```

```python
columnlist =[1, 2 ,3 ,10, 12 ,14,15]
datalist = []
df_industrialfinal = df_industrial[df_industrial['Column_Order'] == 13].drop(columns=['Value','Column_Order','Description']).reset_index()
for ind in columnlist:
    df1 = df_industrial[df_industrial['Column_Order'] == ind]
    df_industrialfinal[str(df1['Description'].iloc[0])] = list(df1['Value'])

df_industrialfinal['date'] = [int(i) for i in df_industrialfinal['YYYYMM']]
df_industrialfinal['Total Energy'] = [float(i) for i in df_industrialfinal['Total Energy Consumed by the Industrial Sector']]
df_industrialfinal

```

```python
columnlist =[1, 2 ,3 ,10, 12 ,13,14,15]
datalist = []
df_commercialfinal = df_commercial[df_commercial['Column_Order'] == 13].drop(columns=['Value','Column_Order','Description']).reset_index()
for ind in columnlist:
    df1 = df_commercial[df_commercial['Column_Order'] == ind]
    df_commercialfinal[str(df1['Description'].iloc[0])] = list(df1['Value'])

df_commercialfinal['date'] = [int(i) for i in df_commercialfinal['YYYYMM']]
df_commercialfinal['Total Energy'] = [float(i) for i in df_commercialfinal['Total Energy Consumed by the Commercial Sector']]
df_commercialfinal
    
```

```python
df_residentfinal.to_csv('residental_energy.csv')
df_commercialfinal.to_csv('Commercial_energy.csv')
df_industrialfinal.to_csv('Industrial_final.csv')
```

```python
df_residentfinal['datetime'] = pd.to_datetime(df_residentfinal['date'], format='%Y%m', errors='coerce')
df_commercialfinal['datetime'] = pd.to_datetime(df_commercialfinal['date'], format='%Y%m', errors='coerce')
df_industrialfinal['datetime'] = pd.to_datetime(df_industrialfinal['date'], format='%Y%m', errors='coerce')

```

```python
plt.plot(df_residentfinal['datetime'],[float(i) for i in df_residentfinal['Total Energy Consumed by the Residential Sector']])
```

```python
df_yearly_resident = df_residentfinal.groupby(df_residentfinal['datetime'].dt.year)['Total Energy'].sum().to_frame(name = 'Resident Total Energy').reset_index()
df_yearly_commercial = df_commercialfinal.groupby(df_commercialfinal['datetime'].dt.year)['Total Energy'].sum().to_frame(name = 'Commercial Total Energy').reset_index()
df_yearly_industrial = df_industrialfinal.groupby(df_industrialfinal['datetime'].dt.year)['Total Energy'].sum().to_frame(name = 'Industrial Total Energy').reset_index()


df_yearly_resident
```

```python
import matplotlib_inline
# get higher quality plots
matplotlib_inline.backend_inline.set_matplotlib_formats('retina')

fig, axs = plt.subplots(1, 3, figsize=(20, 5), tight_layout=True)
for ax, df in zip(axs, [df_yearly_resident, df_yearly_commercial, df_yearly_industrial]):
    ax.plot(df['datetime'],df[df.columns[1]])
    ax.set_xlabel('Year')
    ax.set_ylabel('Total Energy Consumption (Trillion Btu)')
    ax.set_title(df.columns[1])
    plt.rcParams["axes.edgecolor"] = "black"
    plt.rcParams["axes.linewidth"] = 1.5

fig.suptitle('Analyzing Energy Usage in Resident, Commerical, and Industrial', fontsize=25)
plt.savefig('Analyzing Energy Usage in Resident, Commerical, and Industrial.png')
```
